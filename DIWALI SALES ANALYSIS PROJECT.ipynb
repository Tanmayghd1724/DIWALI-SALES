```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hp\Downloads\Diwali Sales Data.csv' , encoding = 'unicode_escape' )
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#drop unrealted column
df.drop(['Status' , 'unnamed1'] , axis = 1 , inplace = True)
```

```python
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#Change Data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column name 
df.rename(column = {'Marital_Status' : 'Shaddi'})
```

```python
#describe() method return description of the data in the DataFrame (i.e. count , mean , std , etc)
df.describe()
```

```python
#use describe() for specific columns 
df[['Age' , 'Orders' , 'Amount']].describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
```

```python
#From the above graph we can see that most of the buyers are female and even the purchasing power of female is greater than men
```

```python
ax = sns.countplot(data = df , x = 'Age Group' , hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False)

sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age)
```

```python
#from the above graph we can see that most of the buyers are from the age group 26-35 years female
```

```python
sales_state = df.groupby(['State'] , as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,6)})
sns.barplot(data = sales_state , x = 'State' , y = 'Orders')
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,6)})
sns.barplot(data = sales_state , x = 'State' , y = 'Amount')
```

```python
#from the abov graph we can see that unexpectedly most of the orders are from  uttar pradesh, maharshtra and karnataka respectively bu\t total/sales is from UP < Karnataka and maharshtra
```

```python
ax = sns.countplot(data = df , x = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['Marital_Status' , 'Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc = {'figure.figsize':(6,4)})
sns.barplot(data = sales_state , x = 'Marital_Status'  , y = 'Amount' , hue = 'Gender')
```

```python
#from the above graph we can see that most of the buyers are married women and they have high purchasing power
```

```python
sns.set(rc = {'figure.figsize':(20,6)})

ax = sns.countplot(data = df , x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,6)})
sns.barplot(data = sales_state , x = 'Occupation' , y = 'Amount')
```

```python
#from the above column we can see that most of the buyers are from IT , Aviation and Health care sector
```

```python
sns.set(rc = {'figure.figsize':(25,6)})

ax = sns.countplot(data = df , x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,6)})
sns.barplot(data = sales_state , x = 'Product_Category' , y = 'Amount')
```

```python
#from the baove graph we can see that most of sold products are from food , clothing  and electronic category
```

```python
sales_state = df.groupby(['Product_ID'] , as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,6)})
sns.barplot(data = sales_state , x = 'Product_ID' , y = 'Orders')
```

```python
#CONCLUSION

#Married women age group between 26-35 from UP , Maharshtra and Karnataka working in IT , Healthcare and Aviation arec more likely to buy productrs from food , clothing  and electronics category
```
